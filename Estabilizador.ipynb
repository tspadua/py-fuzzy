```python
#imports
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl

#Variaveis de Entrada (Antecedent)
var_1 = ctrl.Antecedent(np.arange(-30, 30, 0.5), 'var_1')
var_2 = ctrl.Antecedent(np.arange(-6, 6, 0.1), 'var_2')
var_3 = ctrl.Antecedent(np.arange(-0.4, 0.4, 0.05), 'var_3')
var_4 = ctrl.Antecedent(np.arange(-1, 1, 0.1), 'var_4')

#Variaveis de saída (Consequent)
vector = ctrl.Consequent(np.arange(-6, 6, 0.2), 'vector')

# atribuicao sem o automf
var_1['nvb'] = fuzz.trapmf(var_1.universe, [-30, -30, -18, -12])
var_1['nb'] = fuzz.trimf(var_1.universe, [-16.5, -10.5, -4.5])
var_1['n'] = fuzz.trimf(var_1.universe, [-9, -4.5, 0])
var_1['zo'] = fuzz.trimf(var_1.universe, [-3, 0, 3])
var_1['p'] = fuzz.trimf(var_1.universe, [0, 4.5, 9])
var_1['pb'] = fuzz.trimf(var_1.universe, [4.5, 10.5, 16.5])
var_1['pvb'] = fuzz.trapmf(var_1.universe, [12, 18, 30, 30])
var_2['nb'] = fuzz.trapmf(var_2.universe, [-6, -6, -4.2, -1.7])
var_2['n'] = fuzz.trimf(var_2.universe, [-3.6, -1.7, 0])
var_2['zo'] = fuzz.trimf(var_2.universe, [-1.7, 0, 1.7])
var_2['p'] = fuzz.trimf(var_2.universe, [0, 1.7, 3.6])   
var_2['pb'] = fuzz.trapmf(var_2.universe, [1.7, 4.2, 6, 6])
var_3['nbig'] = fuzz.trapmf(var_3.universe, [-0.4, -0.4, -0.3, -0.15])
var_3['neg'] = fuzz.trimf(var_3.universe, [-0.3, -0.15, 0])
var_3['z'] = fuzz.trimf(var_3.universe, [-0.15, 0, 0.15])
var_3['pos'] = fuzz.trimf(var_3.universe, [0, 0.15, 0.3])   
var_3['pbig'] = fuzz.trapmf(var_3.universe, [0.15, 0.3, 0.4, 0.4])
var_4['neg'] = fuzz.trapmf(var_4.universe, [-1, -1, -0.1, 0])
var_4['z'] = fuzz.trimf(var_4.universe, [-0.1, 0, 0.1]) 
var_4['pos'] = fuzz.trapmf(var_4.universe, [0, 0.1, 1, 1])
vector['nvvb'] = fuzz.trapmf(vector.universe, [-6, -6, -4.8, -3.6])
vector['nvb'] = fuzz.trimf(vector.universe, [-4.8, -3.6, -2.4])
vector['nb'] = fuzz.trimf(vector.universe, [-3.6, -2.4, -1.2])
vector['n'] = fuzz.trimf(vector.universe, [-2.4, -1.2, 0])
vector['z'] = fuzz.trimf(vector.universe, [-1.2, 0, 1.2])
vector['p'] = fuzz.trimf(vector.universe, [0, 1.2, 2.4])
vector['pb'] = fuzz.trimf(vector.universe, [1.2, 2.4, 3.6])
vector['pvb'] = fuzz.trimf(vector.universe, [2.4, 3.6, 4.8])
vector['pvvb'] = fuzz.trapmf(vector.universe, [3.6, 4.8, 6, 6])

var_1.view()
var_2.view()
var_3.view()
var_4.view()
vector.view()

rules = []
rules.append(ctrl.Rule(var_3['nbig'] & var_4['neg'], vector['pvvb']))
rules.append(ctrl.Rule(var_3['neg'] & var_4['neg'], vector['pvb']))
rules.append(ctrl.Rule(var_3['z'] & var_4['neg'], vector['pb']))
rules.append(ctrl.Rule(var_3['z'] & var_4['z'], vector['z']))
rules.append(ctrl.Rule(var_3['z'] & var_4['pos'], vector['nb']))
rules.append(ctrl.Rule(var_3['pos'] & var_4['pos'], vector['nvb']))
rules.append(ctrl.Rule(var_3['pbig'] & var_4['pos'], vector['nvvb']))
rules.append(ctrl.Rule(var_1['nvb'] & var_2['nb'], vector['nvvb']))
rules.append(ctrl.Rule(var_1['nvb'] & var_2['n'], vector['nvvb']))
rules.append(ctrl.Rule(var_1['nvb'] & var_2['zo'], vector['nvb']))
rules.append(ctrl.Rule(var_1['nvb'] & var_2['p'], vector['nb']))
rules.append(ctrl.Rule(var_1['nvb'] & var_2['pb'], vector['n']))
rules.append(ctrl.Rule(var_1['nb'] & var_2['nb'], vector['nvvb']))
rules.append(ctrl.Rule(var_1['nb'] & var_2['n'], vector['nvb']))
rules.append(ctrl.Rule(var_1['nb'] & var_2['zo'], vector['nb']))
rules.append(ctrl.Rule(var_1['nb'] & var_2['p'], vector['n']))
rules.append(ctrl.Rule(var_1['nb'] & var_2['pb'], vector['z']))
rules.append(ctrl.Rule(var_1['n'] & var_2['nb'], vector['nvb']))
rules.append(ctrl.Rule(var_1['n'] & var_2['n'], vector['nb']))
rules.append(ctrl.Rule(var_1['n'] & var_2['zo'], vector['n']))
rules.append(ctrl.Rule(var_1['n'] & var_2['p'], vector['z']))
rules.append(ctrl.Rule(var_1['n'] & var_2['pb'], vector['p']))
rules.append(ctrl.Rule(var_1['zo'] & var_2['nb'], vector['nb']))
rules.append(ctrl.Rule(var_1['zo'] & var_2['n'], vector['n']))
rules.append(ctrl.Rule(var_1['zo'] & var_2['zo'], vector['z']))
rules.append(ctrl.Rule(var_1['zo'] & var_2['p'], vector['p']))
rules.append(ctrl.Rule(var_1['zo'] & var_2['pb'], vector['pb']))
rules.append(ctrl.Rule(var_1['p'] & var_2['nb'], vector['n']))
rules.append(ctrl.Rule(var_1['p'] & var_2['n'], vector['z']))
rules.append(ctrl.Rule(var_1['p'] & var_2['zo'], vector['p']))
rules.append(ctrl.Rule(var_1['p'] & var_2['p'], vector['pb']))
rules.append(ctrl.Rule(var_1['p'] & var_2['pb'], vector['pvb']))
rules.append(ctrl.Rule(var_1['pb'] & var_2['nb'], vector['z']))
rules.append(ctrl.Rule(var_1['pb'] & var_2['n'], vector['p']))
rules.append(ctrl.Rule(var_1['pb'] & var_2['zo'], vector['pb']))
rules.append(ctrl.Rule(var_1['pb'] & var_2['p'], vector['pvb']))
rules.append(ctrl.Rule(var_1['pb'] & var_2['pb'], vector['pvvb']))
rules.append(ctrl.Rule(var_1['pvb'] & var_2['nb'], vector['p']))
rules.append(ctrl.Rule(var_1['pvb'] & var_2['n'], vector['pb']))
rules.append(ctrl.Rule(var_1['pvb'] & var_2['zo'], vector['pvb']))
rules.append(ctrl.Rule(var_1['pvb'] & var_2['p'], vector['pvvb']))
rules.append(ctrl.Rule(var_1['pvb'] & var_2['pb'], vector['pvvb']))

controlador = ctrl.ControlSystem(rules)

#Simulando
calc = ctrl.ControlSystemSimulation(controlador)

valor_entrada1 = float(input('Entrada 1: '))
valor_entrada2 = float(input('Entrada 2: '))
valor_entrada3 = float(input('Entrada 3: '))
valor_entrada4 = float(input('Entrada 4: '))

calc.input['var_1'] = valor_entrada1
calc.input['var_2'] = valor_entrada2
calc.input['var_3'] = valor_entrada3
calc.input['var_4'] = valor_entrada4
calc.compute()

output = calc.output['vector']

var_1.view(sim=calc)
var_2.view(sim=calc)
var_3.view(sim=calc)
var_4.view(sim=calc)
vector.view(sim=calc)

plt.show()

print("\nEntrada 1: %d \nEntrada 2: %.1f \nEntrada 3: %.1f \nEntrada 4: %.1f \nVector %.1f" %(
        valor_entrada1,
        valor_entrada2,
        valor_entrada3,
        valor_entrada4,
        output))
```
